```python
age_group = [0 , 10 , 20 , 30 , 40 , 50 , 60 , 70]
age_bins  = ['0-10' , '11-20 ' ,'21-30' , '31-40' , ' 41-50' , '51-60' , '61-70']


df['Age_Group'] = pd.cut(df['Age'], bins = age_group , labels=age_bins )
```

```python
ratings_group  = [0 , 1 , 2 ,3 ,4 ,5]
ratings_bins  = ['0-1' , '1-2 ' ,'2-3' , '3-4' , '4-5']




df['ratings_group'] = pd.cut(df['Ratings'] , bins=ratings_group , labels=ratings_bins)




```

```python
amount_group  = [100 , 200 , 300 ,400, 500]
amount_bins = ['100-200', '101-200' , '201-300' , '301-400' ]

df['amount_group'] = pd.cut(df['Amount'], bins = amount_group , labels=amount_bins )
```

```python

total_amount_group = [500 , 1000 , 1500 , 2000 , 2500 , 3000 , 3500 , 4000 , 4500 , 5000]
total_amount_bins = ['0-500' , '501-1000' , '1001-1500' , '1501-2000' , '2001-2500' , '2501-3000' , '3001-3500' , '3501-4000' , '4001-4500' ]

df['total_amount_group'] = pd.cut(df['Total_Amount'] , bins=total_amount_group , labels=total_amount_bins)

```

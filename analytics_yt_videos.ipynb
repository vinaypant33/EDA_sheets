```python
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows' , 10000)
pd.set_option('display.max_columns' , 10000)
```

```python
# Reading the data frame : 
df = pd.read_csv("Trending Youtube Videos 2019 to 2020.csv")
```

```python
df.head(10)
```

```python
df['time_checked'] = pd.to_datetime(df['timestamp'])
```

```python
df['year'] = df['time_checked'].apply(lambda x  : x.year)
```

```python
df.head(10)
```

```python
df['month'] = df['time_checked'].apply(lambda x  : x.month)
```

```python
df['month_name'] = df['month'].apply(lambda x  : 'April' if x == 4 else 
                                     'Other' if x == 5 else 
                                     "Name") 
```

```python
def month_change (number):
    if number  ==1 :
        return "January"
    elif number ==2 : 
        return "February"
    elif number == 3:
        return "March"
    elif number == 4 :
        return "April"
    elif number == 5 :
        return "May"
    elif number == 6:
        return "June"
    elif number == 7:
        return "July"
    elif number == 8:
        return "August"
    elif number == 9:
        return "September"
    elif number == 10:
        return "October"
    elif number == 11:
        return "November"
    elif number == 12:
        return "December"
    
    
df['month_name'] = df['month'].apply(lambda x : month_change(x))
```

```python
df.head(20)
```

```python
df = df.drop(['time_checked'] ,axis = 1)
```

```python
df = df.drop(['timestamp'] , axis = 1)
```

```python
df = df.drop(['videostatsid']  ,axis = 1)
```

```python
# Binning for the Amount Total Amount and Age  : 

# Binning Amount :

age_bins = [0,10,20,30,40,50,60,70]
slot  = ['0-10' , '11-20' , '21-30' , '31-40' , '41-50' , '51-60' , '61-70']
df['age_rank'] = pd.cut(df['Age'] , age_bins , labels=slot )
```

```python
df.head()
```

```python
amount_bins  = [0 , 100 , 200 , 300, 400 , 500]
amount_slot = ['0-100' , '101-200' , '201-300'  , '301-400' , '401-500']

df['Amount_rank'] = pd.cut(df['Amount'] , amount_bins , labels = amount_slot)
```

```python
# Bins for the total amount  : 

total_amount_bins  = [0 , 100 , 2000 , 3000 , 4000 , 5000]
total_amount_slot = ['0-1000' , '1001-2000' , '2001-3000' , '3001-4000' , '4001-5000']

df['Total_Amount_rank'] = pd.cut(df['Total_Amount'] , total_amount_bins , labels=total_amount_slot)
```
